```python
def get_fc_trans_kt(hp):
    f=8
    act1,act2a,act2b,act3,act4a,act4b,act5,act6=["linear"]*7+["relu"]
    fc_trans = tf.keras.Sequential()
    
    squeeze=128
    max_params_submodel = 1000000
    
    num_layers_in=hp.Int(name=f"num_layers_in",min_value =1,max_value =8,step=1)
    num_layers_out=hp.Int(name=f"num_layers_out",min_value =1,max_value =8,step=1)
    
    max_params_per_layer = max_params_submodel/(num_layers_in+num_layers_out)
    min_params_per_layer = 10000
    params_between_layers=[input_height*input_width*3*squeeze]
    #params_between_layers=[]
    remaining_params = max_params_submodel - sum(params_between_layers[1:]) 
    remaining_layers = num_layers_in + num_layers_out
    remaining_params_per_layer = remaining_params/remaining_layers
    
    
    fc_trans.add(tf.keras.layers.Flatten(input_shape=(input_height,input_width,3)))
    fc_trans.add(tf.keras.layers.Dense(squeeze,activation=act1,name="squeeze_in"))
    
    units_in_layers=[squeeze]    
    
    ########################
    for layer_num in range(1,remaining_layers+1+1):
        if layer_num <= num_layers_in:
            layer_type = "in"
        if layer_num == num_layers_in+1:
            layer_type = "latent"
        else:
            layer_type = "out"
        
        last_layer_units = units_in_layers[-1]
        
        print(layer_num, remaining_layers,remaining_params_per_layer,last_layer_units,params_between_layers[-1])
        
        if layer_type != "latent":
            try:
                min_v  = last_layer_units*squeeze
                max_v  = int(remaining_params_per_layer)
                step   = last_layer_units
                if max_v > min_v:
                    num_params_for_layer = hp.Int(name=f"in-{layer_num} ll:{last_layer_units}",
                                              min_value = min_v,
                                              max_value = max_v,
                                              step=step)        
                else:
                    raise ValueError(f'min_v: {min_v}, max_v: {max_v}')
            except ValueError as e:
                exit()
                
            num_units_for_layer = max(int(num_params_for_layer/(last_layer_units+1))+1,1)
            
            nm = str(layer_num)+".."+str(num_units_for_layer) + ".."  + str(num_params_for_layer)
            fc_trans.add(tf.keras.layers.Dense(int(num_units_for_layer),name=nm))
            units_in_layers.append(int(num_units_for_layer))

            params_between_layers.append(num_params_for_layer)
            remaining_params = max_params_submodel - sum(params_between_layers[1:])
            if remaining_layers > 0:
                remaining_params_per_layer = remaining_params/remaining_layers
        else:
            fc_trans.add(tf.keras.layers.Dense(squeeze,activation=act1,name="latent"))
            units_in_layers.append(squeeze)
            params_between_layers.append(squeeze*last_layer_units)
            remaining_params = max_params_submodel - sum(params_between_layers[1:])
            remaining_params_per_layer = remaining_params/remaining_layers
            
        remaining_layers -=1
        
    ######################
    fc_trans.add(tf.keras.layers.Dense(squeeze,activation=act1,name="squeeze_out"))    
    fc_trans.add(tf.keras.layers.Dense(input_width*input_height*3,activation=act6))
    fc_trans.add(tf.keras.layers.Reshape((input_height,input_width,3)))
    
    fc_trans.compile(optimizer=tf.keras.optimizers.Adam(
        learning_rate=hp.Float(name="LR",min_value =.000001, max_value = .01, step=.00001)
    ),loss=tf.keras.losses.MSE,metrics="MSE")
    print (fc_trans.summary())
    return fc_trans
    #####################


```

```python
def get_fc_trans():
    f=16
    act1,act2a,act2b,act3,act4a,act4b,act5,act6=["linear"]*7+["relu"]
    fc_trans = tf.keras.Sequential()
    fc_trans.add(tf.keras.layers.Flatten(input_shape=(input_height,input_width,3)))
    fc_trans.add(tf.keras.layers.Dense(max(16*f,64),activation=act1))



    for i in range(1):    
        fc_trans.add(tf.keras.layers.Dense(32*f,activation=act2a))
        fc_trans.add(tf.keras.layers.Dense(16*f,activation=act2b))

    fc_trans.add(tf.keras.layers.Dense(8*f,activation=act3))


    for i in range(1):
        fc_trans.add(tf.keras.layers.Dense(16*f,activation=act4a))
        fc_trans.add(tf.keras.layers.Dense(32*f,activation=act4b))


    fc_trans.add(tf.keras.layers.Dense(max(16*f,64),activation=act5))
    fc_trans.add(tf.keras.layers.Dense(input_width*input_height*3,activation=act6))
    fc_trans.add(tf.keras.layers.Reshape((input_height,input_width,3)))
    
    fc_trans.compile(optimizer=tf.keras.optimizers.Adam(
        learning_rate=.0001),
        loss=tf.keras.losses.MSE,
        metrics="MSE")
    return fc_trans
```

```python
model = get_fc_trans()
```
